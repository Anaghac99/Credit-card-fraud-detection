```python
# next we pass our x data and y data in our train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
# applying machine learning algorithmns
```

```python
#for that first we need to import our machine learning algorithms
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score, precision_score,recall_score
```

```python
#Now we apply Ml algorithmn and check the accuracy score
```

```python
#Now we create a dictionary 'classifier'.Inside this classifier we initialise the which  ML algorithmns using here. here we using logistic regression, and decision tree classifier
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Random Forest Clasifier": RandomForestClassifier()
}

for name,clf in classifier.items(): ##''in this dictionary , first gives the key the value.here 'name' is the key(which type of model it is for ex: logistic regression, decision tree classifier). and classifier is the instance of that
    print(f"\n==========={name}===========")
    clf.fit(x_train,y_train) #using this classifier to fit the data. if it is fit it would pass the x_train and y_train
    y_pred = clf.predict(x_test)#next we prdict the data based on the test data
    print(f"\n\nAccuracy: {accuracy_score(y_test,y_pred)}")
    print(f"\nPrecision Score: {precision_score(y_test,y_pred)}")
    print(f"\nRecall Score: {recall_score(y_test,y_pred)}")
    print(f"\nF1 Score: {f1_score(y_test,y_pred)}")
```

```python
# if the dataset is unbalanced we can only check accuracy score. if all of the values are almost equal then it means that the model is perfect.but here all the values are different. so we need to handle the unbalanced data.
```

```python
#Next we handle the unbalanced data. there are two methods for it over sampling and under sampling
```

```python
#First we apply the under sampling technique
```

```python
legitimate = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
legitimate.shape
```

```python
fraud.shape
```

```python
legitimate_sample = legitimate.sample(n=473)
```

```python
legitimate_sample.shape
```

```python
# to get a new data we add our both legitimate_sample and fraud
```

```python
new_data =pd.concat([legitimate_sample,fraud],ignore_index=True)
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
new_data['Class'].value_counts()
```

```python
# the major problem of under sampling is that we lost lots of data
```

```python
#Once again we divide our data into train,test and split . Before that we make it into dependent and independent features.
```

```python
x= new_data.drop(['Class'],axis=1)
y= new_data['Class']
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Random Forest Clasifier": RandomForestClassifier()
}

for name,clf in classifier.items(): ##''in this dictionary , first gives the key the value.here 'name' is the key(which type of model it is for ex: logistic regression, decision tree classifier). and classifier is the instance of that
    print(f"\n==========={name}===========")
    clf.fit(x_train,y_train) #using this classifier to fit the data. if it is fit it would pass the x_train and y_train
    y_pred = clf.predict(x_test)#next we prdict the data based on the test data
    print(f"\n\nAccuracy: {accuracy_score(y_test,y_pred)}")
    print(f"\nPrecision Score: {precision_score(y_test,y_pred)}")
    print(f"\nRecall Score: {recall_score(y_test,y_pred)}")
    print(f"\nF1 Score: {f1_score(y_test,y_pred)}")
```

```python
#After doing the under sampling , we got a better accuracy score, precision score , recall score and f1 score which are almost approximate than before.
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#next we use smote, after that we can  fit and resample  not transform because we have only 473 data and we want it to resample it to x and y
x_res,y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
x_train, x_test, y_train, y_test =train_test_split(x_res,y_res,train_size=0.2,random_state=42)
```

```python
classifier = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Random Forest Clasifier": RandomForestClassifier()
}

for name,clf in classifier.items(): ##''in this dictionary , first gives the key the value.here 'name' is the key(which type of model it is for ex: logistic regression, decision tree classifier). and classifier is the instance of that
    print(f"\n==========={name}===========")
    clf.fit(x_train,y_train) #using this classifier to fit the data. if it is fit it would pass the x_train and y_train
    y_pred = clf.predict(x_test)#next we prdict the data based on the test data
    print(f"\n\nAccuracy: {accuracy_score(y_test,y_pred)}")
    print(f"\nPrecision Score: {precision_score(y_test,y_pred)}")
    print(f"\nRecall Score: {recall_score(y_test,y_pred)}")
    print(f"\nF1 Score: {f1_score(y_test,y_pred)}")
```

```python
#next we save decision tree classifier model
```
